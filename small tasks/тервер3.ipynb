```python
def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    """
    prior_A: априорная вероятность P(A)
    likelihood_B_given_A: вероятность P(B|A)
    prior_not_A: вероятность P(¬A)
    likelihood_B_given_not_A: вероятность P(B|¬A)
    """
    posterior_A_given_B = (likelihood_B_given_A * prior_A) / (
        (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)
    )
    return posterior_A_given_B
#это в итоге не использую
```
