```python
import pandas as pd

# Загружаем данные
file_path = "cars.csv"
df = pd.read_csv(file_path)
```

```python
from scipy import stats
from statsmodels import stats 
```

```python
df['Brand'] = df['CarName'].str.split().str[0]
```

```python
# Подсчёт уникальных брендов
unique_brands_count = df['Brand'].nunique()

# Вывод результата
print(f"Количество уникальных брендов: {unique_brands_count}")
```

```python
print("Список уникальных брендов:")
print(df['Brand'].unique())
```

```python
corrections = {
    "maxda": "mazda",
    "vokswagen": "volkswagen",
    "porcshce": "porsche",
    "toyouta": "toyota",
    "vw": "volkswagen"}
df["Brand"] = df["Brand"].replace(corrections)
df["Brand"] = df["Brand"].str.capitalize()
# Проверка уникальных значений
print("Уникальные бренды после исправлений:")
print(df["Brand"].unique())
```

```python
# Количество уникальных значений
print(f"Количество уникальных брендов: {df['Brand'].nunique()}")
```
